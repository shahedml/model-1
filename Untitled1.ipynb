```python
# First, let's import the tools we'll need for our project
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# independent variable (years of Experience)
x = np.array([1,2,3,4,5,6,8,9,10]).reshape(-1,1)

# Dependent variable(Salary in thousands of $)
y = np.array([45,50,60,65,70,78,90,95,105])

# Let's print our data to see it
print("Years of Experience(x):")
print(x.flatten())
print("/nSalary in thousands(y)")
print(y)
```

```python
# Create a scatter plot
plt.figure(figsize=(8,6))
plt.scatter(x,y,color='green',label='Actual Data Points')
plt.title('Years of Experience vs. Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (in thouands of $)')
plt.grid(True)
plt.legend()
plt.show
```

```python
#1. 
model = LinearRegression()
#2. Train the model using our data()
model.fit(x,y)
print("Model training complete")
```

```python
# Get the intercept (c)
intercept = model.intercept_
# Get the slope(m)
slope = model.coef_[0]
print(f"The intercept (c) is: {intercept:.2f}")
print(f"The slope (m) is: {slope:.2f}")
```

```python
# Create a scatter plot
plt.figure(figsize=(8,6))
plt.scatter(x,y,color='blue',label='Actual Data Points')
# plot the original data points
# we use model.predict(x) to get the y-values that lie on line
plt.plot(x, model.predict(x), color='red' ,linewidth=2, label='Best-Fit Line')

plt.title('Years of Experience vs. Salary with Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (in thouands of $)')
plt.grid(True)
plt.legend()
plt.show
```

```python
# let's predict the salary for a person with 7 years of experience
years_of_experience= 7
#we need to put it into a 2D array formet for the model
new_x= np.array([[years_of_experience]])
#use the model to predict
predicted_salary =model.predict(new_x)
print((f"predicted salary for {years_of_experience} years of experience: $ {predicted_salary[0]:.2f}k"))
```
