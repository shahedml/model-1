```python
# calculator programing 

a = float(input("enter first number:"))
b = float(input("enter second number:"))
operator = input("+,-,*,/,**,%):")

if operator == "+":
    print(a+b)
elif operator == "-":
    print(a-b)
elif operator == "*":
    print(a*b)
elif operator == "/":
    print(a/b)
elif operator == "**":
    print(a**b)
elif operator == "%":
    print(a%b)
else:
    print("not")
    
```
